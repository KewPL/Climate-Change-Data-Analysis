```python
%%time
#Measuring the time of creating the heatmap
corr = datasummary.corr() #Selecting the appropriate database which I have imported

#Customising the font
font1 = {'family':'serif','color':'red','size':20}
font2 = {'family':'serif','color':'black','size':15}

fig, ax = plt.subplots(figsize = (20,6)) #This is the command used for changing the size of the chart(20 is the horizontal lenght and 6 is the height)
g= sns.heatmap(corr,ax=ax, annot= True, cmap='Greens') #Command used to state the positions of the data and 'True' determines that the data has captions(text inside the boxes), also used to detemrine the colour of the heatmap

plt.title('The difference of the summary value between the variables', fontdict = font1) #Inserting a title for the graph
plt.xlabel("Type of data", fontdict = font2)
plt.ylabel("Type of data", fontdict = font2)
```
